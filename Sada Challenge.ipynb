```python
import pandas as pd
import numpy as np
```

```python
# Download the data file at https://gist.github.com/bobbae/b4eec5b5cb0263e7e3e63a6806d045f2 as "raw".
#read the file into a program memory.

df = pd.read_csv('Raw.csv')
df
```

```python
#Once the data is read into memory in your program, print out how many rows of the data is in the CSV data. 
#This is your first printed answer.

df.info
```

```python
#For each row, there are columns. One of the columns is the "profit". 
#The data in this profit column are not always valid. Sometimes the data is non-numeric. 
#Remove all rows with 'profit' that is not numerical value. Then print out how many rows of data are left.
#after removing all the rows with invalid non-numeric profit column data. This is your second printed answer.

data2 = df[pd.to_numeric(df['Profit (in millions)'], errors ='coerce').notnull()] 
data2
```

```python
#You can now convert the content your read into memory in your program in Part 1 into JSON format 
#Write it out to another file called data2.json which should only contain rows of data that have the valid numeric profit values.
#Each row in the Array should contain data columns like year, rank, company, revenue, and profit.

data2.to_json('data2.json')
df = pd.read_json('data2.json')
df
```

```python
#Order the data based on the profit value. Print the top 20 rows with the highest profit values. 
#This is your third printed answer.

data = df.sort_index(axis=0,ascending=False) 
data.head(20)
```
